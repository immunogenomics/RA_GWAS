# Code used to create Figure 1

```r
library(data.table)
library(ggplot2)
library(RColorBrewer)
library(ggsci)
library(magrittr) 
```

## Figure 1b

```r
#load data
load("../data/Figure1_pca.RData")
df <- out

#color code
pal_npg("nrc")(6) -> colpal

#plot
options(repr.plot.width=5, repr.plot.height=4)

subset(df,type=="KG") %>% ggplot(aes(x=PC1, y=PC2, color=Pop )) + #shape=type, size=type
    geom_point(size=0.01) +
    theme_bw() +
    scale_color_manual(values = c(
      'EUR' = colpal[1],
      'EAS' = colpal[2],
      'AFR' = colpal[3],
      'SAS' = colpal[4],
      'ARB' = colpal[5],
      'AMR' = colpal[6]),
      #'ARB:1' = colpal[5],
      #'ARB:2' = colpal[6],
      #'ARB:3' = colpal[7],
      #'AMR' = colpal[8]),
      name="") +
    labs(title="1KG only") +
    scale_x_continuous(limits=c(-0.02, 0.04) ) +
    scale_y_continuous(limits=c(-0.04, 0.04) ) +
    #scale_shape_manual(values=c(20,5),name="") +
    #scale_size_manual(values=c(1,5),name="") +
    theme( axis.line = element_line(colour = "black"), axis.text=element_text(colour="black") ) +
    theme( panel.grid.major = element_blank(), panel.grid.minor = element_blank() ) +
    theme( panel.border = element_blank(), panel.background = element_blank() ) 

subset(df,type=="GWAS") %>% ggplot(aes(x=PC1, y=PC2, color=Pop )) + #shape=type, size=type
    geom_point(size=0.01) +
    theme_bw() +
    scale_color_manual(values = c(
      'EUR' = colpal[1],
      'EAS' = colpal[2],
      'AFR' = colpal[3],
      'SAS' = colpal[4],
      'ARB' = colpal[5],
      'AMR' = colpal[6]),
      #'ARB:1' = colpal[5],
      #'ARB:2' = colpal[6],
      #'ARB:3' = colpal[7],
      #'AMR' = colpal[8]),
      name="") +
    labs(title="GWAS only") +
    scale_x_continuous(limits=c(-0.02, 0.04) ) +
    scale_y_continuous(limits=c(-0.04, 0.04) ) +
    #scale_shape_manual(values=c(20,5),name="") +
    #scale_size_manual(values=c(1,5),name="") +
    theme( axis.line = element_line(colour = "black"), axis.text=element_text(colour="black") ) +
    theme( panel.grid.major = element_blank(), panel.grid.minor = element_blank() ) +
    theme( panel.border = element_blank(), panel.background = element_blank() ) 
```

## Figure 1c

```r
#load data
load("../data/Figure1_umap.RData")
df3 <- out

#set level
df3$Group <- factor(df3$Group,
   levels=c(
       "US","UK","Spain","Sweden","Netherland","Canada","France",
       "Japanese","Chinese","Korean",
       "African","Malaysians","Arab"
))

#plot
options(repr.plot.width=5.5, repr.plot.height=4)

df3[sample(row.names(df3)),] %>% ggplot(aes(x=UMAP1, y=UMAP2, color=Group )) + #shape=type, size=type
    geom_point(size=0.01) +
    theme_bw() +
    scale_color_d3("category20",name="") +
    labs(title="all gwas") +
    scale_x_continuous(limits=c(-50, 10) ) +
    scale_y_continuous(limits=c(-10, 12) ) +
    #scale_shape_manual(values=c(20,5),name="") +
    #scale_size_manual(values=c(1,5),name="") +
    theme( axis.line = element_line(colour = "black"), axis.text=element_text(colour="black") ) +
    theme( panel.grid.major = element_blank(), panel.grid.minor = element_blank() ) +
    theme( panel.border = element_blank(), panel.background = element_blank() ) 
    #theme(strip.background = element_blank() ) +
    #theme(strip.text = element_text( size = 12, color = "black"))
    #looks good, use this
```
